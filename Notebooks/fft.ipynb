```python
import numpy as np
import matplotlib.pyplot as plt

N = 44100 # number of sample points
L = 2000 # number of points to be shown

sampling_freq = 44100 
rec_time = (N-1) / sampling_freq

data = np.loadtxt("sound.csv", delimiter=",").flatten()
data = np.trim_zeros(data, 'b')

x = np.linspace(0, L, L)

fig, ax = plt.subplots(figsize=(20,10))
ax.plot(x, data[:L])
ax.set(title = "First " + str(L) + " samples of an audio signal",
       xlabel = "Sample number")
fig.savefig("Visuals/wave.png")
```
